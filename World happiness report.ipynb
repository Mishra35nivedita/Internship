```python
# We can clearly see that are outliers present in the data of few columns. Let's remove the outliers
from scipy.stats import zscore
```

```python
z_score=zscore(data[['Family','Generosity','Trust (Government Corruption)','Dystopia Residual']])
abs_z_score= np.abs(z_score)
```

```python
#We can see some positive skewness in the data. Removing skewness

X=data.drop(columns=['Happiness Score','Region'], axis=1)
y=data['Happiness Score']
```

```python
X
```

```python
from sklearn.preprocessing import power_transform
X=power_transform(X,method='yeo-johnson')

```

```python
data_x=pd.DataFrame(data=X)
```

```python
data_x.skew()
```

```python
#Standardizing data
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#The skewness has been fixed now. Splitiing data in train and test-
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_scaled,y, random_state=50, test_size=0.30)
```

```python
X_train
```

```python
#Model instantiating and training
#Using Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
pred_train=reg.predict(X_train)
pred_test=reg.predict(X_test)
print(f" training accuracy is {r2_score(y_train,pred_train)}")
print(f" testing accuracy is {r2_score(y_test,pred_test)}")
```

```python
mean_score = metrics.mean_absolute_error(y_test,pred_test)
print("Mean Absolute Error Score:",mean_score)
```

```python
#Checking cross validation score to ensure model isn't overfitting-
from sklearn.model_selection import cross_val_score
```

```python
for k in range(3,6):
    cv_score=cross_val_score(reg,X,y,cv=k)
    cv_mean=cv_score.mean()
    print(f" at {k}, cross val score is {cv_mean}")
    print("\n")
```

```python
#At cv=3, we are getting cross val score 89.5% and test accuracy is 99.1%. Let's use hyperperameter tuning technique to stabilize model and generalize it for unseen data.
```
