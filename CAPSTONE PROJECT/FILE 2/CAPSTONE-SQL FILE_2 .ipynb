```python
# Importing Libraries
import pandas as pd
import sqlite3
```

```python
#Loading dataset
df= pd.read_csv('clothing_b.csv')
df.head()
```

```python
df['Price'] = df['Price'].str.replace('₹ ', '', regex=True)
```

```python
df
```

```python
# Creating Table_1.csv
data1_df = df[['Sno', 'Brand', 'Category', 'Product_type', 'Product_Name', 'Product_code', 'Price']]
data1_df.to_csv('Table_1.csv', index=False)

# Creating Table_2.csv
data2_df = df[['Sno', 'Wear_type', 'Material_type', 'Reviews', 'color1', 'color2']]
data2_df.to_csv('Table_2.csv', index=False)

# Creating Table_3.csv
data3_df = df[['Sno','Country_of_origin', 'Return_time', 'Size1', 'Size2', 'Size3', 'Size4','Occassion']]
data3_df.to_csv('Table_3.csv', index=False)
```

```python
# Creating SQLite database 
conn = sqlite3.connect('clothing_b.db')
cursor = conn.cursor()
```

Creating 3 tables inside DataBase clothing_b named as data1, data2 and data3

```python
# Create data1 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data1 (
        Sno INTEGER PRIMARY KEY,
        Brand TEXT,
        Category TEXT,
        Product_type TEXT,
        Product_Name INTEGER,
        Product_code TEXT
        Price REAL
        
    )
''')

# Create data2 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data2 (
        Sno INTEGER PRIMARY KEY,
        Wear_type TEXT,
        Material_type TEXT,
        Reviews TEXT
        color1  TEXT,
        color2  TEXT
    )
''')

# Create data3 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data3 (
        Sno INTEGER PRIMARY KEY,
        Country_of_origin TEXT,
        Return_time INTEGER,
        Size1 TEXT,
        Size2 TEXT,
        Size3 TEXT,
        Size4 TEXT,
        Occassion TEXT
    )
''')

# Commit the changes 
conn.commit()


print("SQLite database 'clothing_b' with tables 'data1,' 'data2,' and 'data3' created successfully.")
```

```python
# Paths to CSV files
data1_csv = 'Table_1.csv'
data2_csv = 'Table_2.csv'
data3_csv = 'Table_3.csv'

# Read CSV files into DataFrames
data1_df = pd.read_csv(data1_csv)
data2_df = pd.read_csv(data2_csv)
data3_df = pd.read_csv(data3_csv)

# Uploading data to respective tables
data1_df.to_sql('data1', conn, if_exists='replace', index=False)
data2_df.to_sql('data2', conn, if_exists='replace', index=False)
data3_df.to_sql('data3', conn, if_exists='replace', index=False)

# Commit changes 
conn.commit()

print("Data uploaded to their respective tables in the 'clothing_b' database")
```

QUERIES FOR TABLE 1

```python
#1. Writing a query to Retrieve the Top 20  Most Expensive Products

query = ''' 
SELECT Product_Name, Price
FROM Data1
order by Price desc limit 20; 
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#2. Writing a query to Calculate the Total Price for Each :

query = ''' 
SELECT Category, SUM(Price) AS Total_Price
FROM Data1
GROUP BY Category; 
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#3. Writing a query to Retrieve Products with a Price Range 2000 to 5000

query = ''' 
SELECT Product_Name, Price 
FROM Data1
WHERE Price BETWEEN 200 AND 5000; 
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#4. Writing a query to Calculate the Average Price for Each Brand

query = ''' 
SELECT Brand, AVG(Price) AS Average_Price
FROM Data1
GROUP BY Brand;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#5. Writing a query to find Products with Unique Brand-Category Combinations

query = ''' 
SELECT Brand, Category, Product_Name
FROM Data1
GROUP BY Brand, Category;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#5. Writing a query to Calculate the Total Price for Each Material Type.

query = ''' 
SELECT Data2.Material_type, SUM(Data1.Price) AS Total_Price
FROM Data1
INNER JOIN Data2 ON Data1.Sno = Data2.Sno
GROUP BY Data2.Material_type;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#3. Writing a query to Retrieve the Top 30 Most Expensive Products by Weartype.

query = ''' 
SELECT Data2.Wear_type, Data1.Price
FROM Data1
JOIN Data2 ON Data1.Sno = Data2.Sno
ORDER BY Data1.Price DESC
LIMIT 30;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#1 Writing a query to find Brandname, Category , Producttype ,ProductName  with a listed price greater than the average listed price in table 1.

query = ''' 
SELECT Brand, Category, Product_type, Product_Name, Price
FROM Data1
WHERE Price > (SELECT AVG(Price) FROM Data1);


'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#2. Writing a query to list all Brandname, Category , Producttype ,ProductName  along with their corresponding color information from Table2

query = ''' 
SELECT 
    Data1.Brand AS Brandname,
    Data1.Category,
    Data1.Product_type AS Producttype,
    Data1.Product_Name AS ProductName,
    Data2.color1,
    Data2.color2
FROM 
    Data1
JOIN 
    Data2 ON Data1.Sno = Data2.Sno;

'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#3. Writing a query to  find the average reviews for each Producttype ,ProductName using table 1 and table 2
query = ''' 
SELECT 
    Data1.Product_type,
    Data1.Product_Name,
    AVG(Data2.Reviews) AS Average_Reviews
FROM 
    Data1
JOIN 
    Data2 ON Data1.Sno = Data2.Sno
GROUP BY 
    Data1.Product_type,
    Data1.Product_Name;

'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#4. Writing a query to find products with a Product name, material type that matches the most common material type in (use table1 and table 2)
query = ''' 
SELECT Data1.Product_Name, Data2.Material_type
FROM Data1 
JOIN Data2 ON Data1.Sno = Data2.Sno
JOIN (
    SELECT Material_type, COUNT(*) AS count_of_material
    FROM Data2
    GROUP BY Material_type
    ORDER BY count_of_material DESC
    ) AS most_common_material ON Data2.Material_type = most_common_material.Material_type;

'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#5. Writing a query to  list all products  Weartype , Material type  along with their corresponding sizes from   (use table 2 and table 3)
query = ''' 
SELECT
    Data2.Wear_type,
    Data2.Material_type,
    Data3.Size1,
    Data3.Size2,
    Data3.Size3,
    Data3.Size4
FROM
    Data2
JOIN
    Data3
ON
    Data2.Sno = Data3.Sno
    LIMIT 610;



'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```

```python
#6. Writing a query to find the average return time for each product type (use table1 and table 3)
query = ''' 
SELECT 
    Data1.Product_type,
    AVG(Data3.Return_time) AS Average_Return_Time
FROM 
    Data1
JOIN 
    Data3 ON Data1.Sno = Data3.Sno
GROUP BY 
    Data1.Product_type;



'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)


# Print the DataFrame
result_df
```
