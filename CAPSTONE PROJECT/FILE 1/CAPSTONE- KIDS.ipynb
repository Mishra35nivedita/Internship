```python
#Searching for kid's apparel
search1=driver.find_element(By.XPATH,'/html/body/main/div[18]/section[1]/a[1]')
search1.click()
```

```python
# Fetching product URL first
product_url=[]
start=0
end=1

for page in range(start,end):
    urls=driver.find_elements(By.XPATH,'//div[@class="li-inner card col-md-12 col-sm-12 col-xs-12 p-0"]/div/div/a[1]')
    for url in urls:
        product_url.append(url.get_attribute('href'))
        print("Appended URL:", url.get_attribute('href'))
    print("Finished processing")
        
```

```python
len(product_url)
```

```python
#Extracting first few records for kids
for url in product_url[0:30]:
    driver.get(url)
    time.sleep(2)
    

Brandname=[]
Category=[]
Producttype=[]
ProductName=[]
Productcode=[]
Price=[]
Weartype=[]
Materialtype=[]
Returntime=[]
reviews=[]
Countryoforigin=[]
reviews=[]
Occasion=[]
color1=[]
color2=[]
Size1=[]
Size2=[]
Size3=[]
Size4=[]
```

```python
for url in product_url[0:30]:
    driver.get(url)
    time.sleep(2)
    
    
    try:
        Brand_name=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/div[2]/div/div[1]")
        Brandname.append(Brand_name.text)
    except NoSuchElementException:
        Brandname.append('--------')
   
    try:
        Cat=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/div[11]/div/h2/button")
        Category.append(Cat.text)
    except NoSuchElementException:
        Category.append('--------')
        

    try:
        Product_type=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/section[1]/div/div[1]/div[2]/div/ul/li[2]/table[10]/tbody/tr/td[2]")
        Producttype.append(Product_type.text)
    except NoSuchElementException:
        Producttype.append('--------')
        
    try:
        Product_Name=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/div[2]/div/div[2]")
        ProductName.append(Product_Name.text)
    except NoSuchElementException:
        ProductName.append("--------")
        
    try:
        Product_code=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/section[1]/div/div[1]/div[2]/div/ul/li[4]/p/span")
        Productcode.append(Product_code.text)
    except NoSuchElementException:
        Productcode.append("--------")
        
    try:
        Price_u=driver.find_element(By.XPATH,"/html/body/main/div[19]/section/div/div[2]/div/div[2]/div/div[2]/div[2]/div/div[3]/div/ul/li/div/ul/li/span[4]")
        Price.append(Price_u.text)
    except NoSuchElementException:
        Price.append("--------")
```

```python
len(Brandname), len(Category), len(Producttype), len(ProductName), len(Productcode), len(Price)
```

```python
data = {
    'Brand': Brandname,
    'Category': Category,
    'Product_type': Producttype,
    'Product_Name': ProductName,
    'Product_code': Productcode,
    'Price': Price
    }
```

```python
df = pd.DataFrame(data)
```

```python
df.rename_axis('Sno',inplace=True)
```
