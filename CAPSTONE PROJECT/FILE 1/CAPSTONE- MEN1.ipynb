```python
#Searching for men's apparel
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[2]/div[2]/div[2]/div/div/div[12]/a/picture/img')
search.click()
```

```python
# Fetching product URL first
product_url=[]
start=0
end=1

for page in range(start,end):
    urls=driver.find_elements(By.XPATH,'//div[@class="product-tile"]/div/div/a[1]')
    for url in urls:
        product_url.append(url.get_attribute('href'))
        print("Appended URL:", url.get_attribute('href'))
    print("Finished processing")
        
```
