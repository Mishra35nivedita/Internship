```python
# Fetching product URL first
product_url=[]
start=0
end=1

for page in range(start,end):
    urls=driver.find_elements(By.XPATH,'//div[@class="MuiGrid-root PlpWeb_product-wrapper__3U7s0 MuiGrid-item MuiGrid-grid-xs-4 MuiGrid-grid-md-4 MuiGrid-grid-lg-4"]/div/div/a')
    for url in urls:
        product_url.append(url.get_attribute('href'))
        print("Appended URL:", url.get_attribute('href'))
    print("Finished processing")
        
```

```python
for url in product_url[0:30]:
    driver.get(url)
    time.sleep(2)
    
    
    try:
        Brand_name=driver.find_element(By.XPATH,"/html/body/div[1]/main/section[3]/footer/div[1]/div/div/div[1]/div/div[1]/p")
        Brandname.append(Brand_name.text)
    except NoSuchElementException:
        Brandname.append('--------')
   
    try:
        Cat=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/div/div/div/div[2]/div[3]/a/span")
        Category.append(Cat.text)
    except NoSuchElementException:
        Category.append('--------')
        

    try:
        Product_type=driver.find_element(By.XPATH,"/html/body/div[1]/main/div[2]/div/div[2]/section/div/div[1]/div[2]/div/div/div/div/ul/li[1]")
        Producttype.append(Product_type.text)
    except NoSuchElementException:
        Producttype.append('--------')
        
    try:
        Product_Name=driver.find_element(By.XPATH,"/html/body/div[1]/main/div[2]/div/div[2]/div[1]/div[3]")
        ProductName.append(Product_Name.text)
    except NoSuchElementException:
        ProductName.append("--------")
        
    try:
        Product_code=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/div/div/div/div[1]")
        Productcode.append(Product_code.text)
    except NoSuchElementException:
        Productcode.append("--------")
        
    try:
        Price_u=driver.find_element(By.XPATH,"/html/body/div[1]/main/div[2]/div/div[2]/div[2]/div[2]/div[2]/s")
        Price.append(Price_u.text)
    except NoSuchElementException:
        Price.append("--------")
```
